```python
import pandas as pd
import numpy as np
```

```python
columns = ['age', 'work_class', 'final_weight', 'education', 'education_id', 'marital_status', 'occupation', 'relationship', 
          'race', 'sex', 'capital_gain', 'capital_loss', 'work_hours_per_week', 'native_country', 'income']
df = pd.read_csv('adult.data', header = None)
df.columns = columns
```

```python
df.info()
```

```python
df
```

#### Converting binary labels to 0-1s

```python
df['income'] = df['income'].apply(lambda x: 1 if '>' in x else 0)
df['sex'] = df['sex'].apply(lambda x: 0 if 'Male' else 1)
```

#### Dropping columns that are uninformative or carry similar information to other columns

```python
df.drop(['education_id', 'final_weight', 'relationship'], axis = 1, inplace = True)
```

```python
df
```

```python
normalise_cols = ['capital_gain', 'capital_loss', 'work_hours_per_week', 'age']
for col in normalise_cols:
    df[col]=(df[col]-df[col].min())/(df[col].max()-df[col].min())
```

```python
dummy_cols = ['work_class', 'education', 'marital_status', 'occupation', 'race', 'native_country']
dummies = pd.get_dummies(df, columns=dummy_cols)
```

#### Saving the cleaned dataframe

```python
dummies.to_csv('adult_cleaned.csv')
```
